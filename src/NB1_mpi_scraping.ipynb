```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')



df1=pd.read_csv('/content/drive/MyDrive/Capstone-Thesis/Datasets/kiva_small_loans/loan_themes_by_region.csv')
print(df1.shape) #shows num of row and column
df1.head(10)
```

```python
# remove columns (Loan Theme ID, Loan Theme Type, forkiva, geocode_old, ISO, number, amount, geocode, geo, lat, lon, mpi_geo)

cols_to_drop = ['Loan Theme ID', 'Loan Theme Type', 'forkiva', 'geocode_old', 'ISO', 'number', 'amount', 'geocode', 'geo', 'lat', 'lon', 'mpi_geo']
df1 = df1.drop(columns=cols_to_drop, errors='ignore')
print(df1.shape)
df1.head(10)
```

```python
# duplicate rows

duplicate_rows_df = df1[df1.duplicated()]
print("Number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
# drop the duplicate rows

df1 = df1.drop_duplicates()
print("Shape of dataframe after removing duplicates: ", df1.shape)
df1.head(10)
```

```python
df2=pd.read_csv('/content/drive/MyDrive/Capstone-Thesis/Datasets/kiva_small_loans/kiva_mpi_region_locations.csv')
print(df2.shape) #shows num of row and column
df2.head(3)
```

```python
print(df1.isnull().sum())
```

```python
# Find null values in df2
print(df2.isnull().sum())
```

```python
# drop the row where LocationName, region  ,MPI   is null

df2 = df2.dropna(subset=['LocationName', 'region', 'MPI'])
print(df2.shape)
df2.head(3)
```

```python
df1 = df1.dropna(subset=['names', 'mpi_region'])
print(df1.shape)
df1.head(3)
```

```python

# Merge df1 and df2 based on mpi_region and LocationName, keeping only the MPI column from df2
merged_df = pd.merge(df1, df2[['LocationName', 'MPI']], left_on='mpi_region', right_on='LocationName', how='left')

print(merged_df.shape)
merged_df.head(10)
```

```python
# Drop specified columns
cols_to_drop = ['Partner ID', 'Field Partner Name', 'sector',  'LocationName_y', 'rural_pct']
merged_df = merged_df.drop(columns=cols_to_drop, errors='ignore')

print(merged_df.shape)


# Find null values
null_values = merged_df.isnull().sum()
null_values
```

```python
merged_df.head(10)
```

```python

# Iterate through rows where 'MPI' is null
for index, row in merged_df[merged_df['MPI'].isnull()].iterrows():
    # Check if any word from 'names' column exists in 'region' column of df2
    name_words = row['names'].split()  # Split 'names' into words
    for word in name_words:
        if any(word in region for region in df2['region']):
            # If a match is found, fill the 'MPI' with a suitable value (e.g., from df2)
            # You might want to refine this logic based on how you want to fill the missing value
            # Here, I'm taking the first matching region's MPI.
            matching_region = df2[df2['region'].str.contains(word, case=False, na=False)].iloc[0]
            merged_df.loc[index, 'MPI'] = matching_region['MPI']
            break  # Exit the inner loop once a match is found
```

```python
# Find null values in the merged dataframe
null_values = merged_df.isnull().sum()
null_values
```

```python
# prompt: if the MPI is null check it the 'region' column == df2's 'region'. if matches then fillup the empty cell then check for null values

# Iterate through rows where 'MPI' is null
for index, row in merged_df[merged_df['MPI'].isnull()].iterrows():
    # Check if the 'region' in df2 matches the 'mpi_region' in the current row
    matching_regions = df2[df2['region'] == row['region']]
    if not matching_regions.empty:
        merged_df.loc[index, 'MPI'] = matching_regions['MPI'].iloc[0]

# Find null values in the merged dataframe
null_values = merged_df.isnull().sum()
null_values
```

```python
# if the MPI is null check it the merged_df's 'LocationName' column == df2's 'LocationName'. if matches then fillup the empty cell then check for null values

# Iterate through rows where 'MPI' is null
for index, row in merged_df[merged_df['MPI'].isnull()].iterrows():
    # Check if 'LocationName' in merged_df matches 'LocationName' in df2
    matching_locations = df2[df2['LocationName'] == row['LocationName_x']]
    if not matching_locations.empty:
        merged_df.loc[index, 'MPI'] = matching_locations['MPI'].iloc[0]

# Find null values in the merged dataframe
null_values = merged_df.isnull().sum()
null_values
```
